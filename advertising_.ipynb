## Data loading

### Subtask:
Load the advertising dataset.


**Reasoning**:
Import pandas and load the advertising dataset into a DataFrame, then print the first 5 rows and the shape of the DataFrame.



```python
import pandas as pd

df_advertising = pd.read_csv('advertising.csv')
display(df_advertising.head())
print(df_advertising.shape)
```

## Data exploration

### Subtask:
Explore the advertising dataset to understand its characteristics.


**Reasoning**:
Explore the dataset by examining its shape, data types, missing values, descriptive statistics, correlation matrix, and data distribution.



```python
# 1. Data Shape
print("Data Shape:", df_advertising.shape)

# 2. Data Types
print("\nData Types:\n", df_advertising.dtypes)

# 3. Missing Values
print("\nMissing Values:\n", df_advertising.isnull().sum())

# 4. Descriptive Statistics
print("\nDescriptive Statistics:\n", df_advertising.describe())

# 5. Correlation Matrix
print("\nCorrelation Matrix:\n", df_advertising.corr())

# 6. Distribution of Data (Histograms)
import matplotlib.pyplot as plt
df_advertising.hist(figsize=(10, 8))
plt.show()
```

## Data visualization

### Subtask:
Visualize the relationship between advertising spending on different media (TV, Radio, Newspaper) and Sales.


**Reasoning**:
Create scatter plots to visualize the relationship between advertising spending on TV, Radio, and Newspaper and Sales, along with regression lines to illustrate the relationship.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot for TV vs Sales
plt.figure(figsize=(8, 6))
sns.regplot(x='TV', y='Sales', data=df_advertising, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.xlabel('TV Advertising Spending')
plt.ylabel('Sales')
plt.title('Relationship between TV Advertising and Sales')
plt.show()

# Scatter plot for Radio vs Sales
plt.figure(figsize=(8, 6))
sns.regplot(x='Radio', y='Sales', data=df_advertising, scatter_kws={'color': 'green'}, line_kws={'color': 'red'})
plt.xlabel('Radio Advertising Spending')
plt.ylabel('Sales')
plt.title('Relationship between Radio Advertising and Sales')
plt.show()

# Scatter plot for Newspaper vs Sales
plt.figure(figsize=(8, 6))
sns.regplot(x='Newspaper', y='Sales', data=df_advertising, scatter_kws={'color': 'orange'}, line_kws={'color': 'red'})
plt.xlabel('Newspaper Advertising Spending')
plt.ylabel('Sales')
plt.title('Relationship between Newspaper Advertising and Sales')
plt.show()
```
